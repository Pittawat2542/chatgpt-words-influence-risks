````python
import json
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
````

````python
with open('data/positive_game_stories.json') as f:
    data = json.load(f)
    del data['count']

    positive_stories = pd.read_json(json.dumps(data['game_stories']))

    img = WordCloud(width=1600, height=800, background_color="white").generate(' '.join(positive_stories['story']).lower())
    plt.figure(figsize=(20, 10), dpi=120)
    plt.imshow(img)
    plt.axis("off")
    plt.savefig('positive_stories.png', bbox_inches='tight')
    print(list(img.words_.keys())[:3])
````

````python
with open('data/negative_game_stories.json') as f:
    data = json.load(f)
    del data['count']

    negative_game_stories = pd.read_json(json.dumps(data['game_stories']))

    img = WordCloud(width=1600, height=800, background_color="white").generate(' '.join(negative_game_stories['story']).lower())
    plt.figure(figsize=(20, 10), dpi=120)
    plt.imshow(img)
    plt.axis("off")
    plt.savefig('negative_stories.png', bbox_inches='tight')
    print(list(img.words_.keys())[:3])
````
